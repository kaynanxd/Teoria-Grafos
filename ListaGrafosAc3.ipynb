Defina o que é:

#####A. O problema de fluxo máximo de s para t em uma rede G(V,A)


R.: O problema de fluxo máximo de s para t em uma rede G(V,A) consiste em encontrar o maior valor de fluxo que pode ser enviado do vértice origem s ao vértice destino t, respeitando as capacidades dos arcos e a conservação de fluxo nos vértices intermediários.


#####B. O fluxo máximo de s para todo t em uma rede G(V,A)

R.: O fluxo máximo de s para todo t em uma rede G(V,A) é o conjunto dos valores máximos de fluxo que podem ser enviados do vértice origem s para cada vértice destino t da rede, respeitando as capacidades dos arcos e a conservação de fluxo.

##### C. Fluxo de corte mínimo

R.: Fluxo de corte mínimo é o conceito associado ao corte s−t de menor capacidade em uma rede. Sua capacidade corresponde ao valor máximo de fluxo que pode ser transportado da origem s ao destino t, conforme o Teorema do Fluxo Máximo e Corte Mínimo.

##### D. As condições para se calcular o fluxo máximo em uma rede G(V,A)

R.: Para calcular o fluxo máximo em uma rede G(V,A), devem ser satisfeitas as seguintes condições:

1. O grafo deve ser dirigido (orientado).
2. Toda capacidade uij deve ser um número inteiro positivo.
3. Não pode existir um caminho direto entre s e t formado por arcos com capacidade infinita.
4. O grafo não deve conter laços (elos).
5. O grafo deve ser simples, ou seja, não pode ser um multigrafo.
6. Deve haver conservação de fluxo em todos os nós da rede, exceto na origem s e no destino t: a quantidade de fluxo que entra em um nó intermediário deve ser igual à quantidade que sai dele.

##### D. Grafo residual GR(V,A)

R.: O grafo residual GR(V,A) é o grafo obtido a partir de uma rede com fluxo, contendo as capacidades residuais dos arcos. Ele indica quanto fluxo adicional ainda pode ser enviado pela rede e é utilizado na busca de caminhos aumentantes para o cálculo do fluxo máximo.

##### F. Arco saturado em uma rede orientada

R.: Um arco saturado em uma rede orientada é um arco cujo fluxo atual é igual à sua capacidade máxima. Assim, sua capacidade residual é zero e nenhum fluxo adicional pode ser enviado por esse arco na direção considerada.

## Questao 3 ✅:
Corte mínimo com o menor número de arcos. Suponha que desejamos identificar entre todos os
cortes mínimos, um corte mínimo contendo o menor número de arcos. Mostre que se trocarmos uij
por u’ij = m uij + 1, o corte mínimo com respeito às capacidades u’ij é um custo mínimo com respeito
às capacidades uij contendo o menor número de arcos.

### Questão 3

Seja \(C\) um corte qualquer. Com a transformação

$$
u'_{ij}=m\,u_{ij}+1
$$

a capacidade do corte torna-se

$$
u'(C)=\sum_{(i,j)\in C}(m\,u_{ij}+1)
      =m\,u(C)+|C|.
$$

onde \(u(C)\) é a capacidade original do corte e \(|C|\) é o número de arcos do corte.

Como \(m\) é suficientemente grande, a minimização de \(u'(C)\) garante primeiro a minimização de \(u(C)\), ou seja, a obtenção de um corte mínimo. Entre cortes com a mesma capacidade mínima, o termo \(|C|\) faz com que seja escolhido aquele com o menor número de arcos.

Logo, o corte mínimo com respeito às capacidades \(u'_{ij}\) é um corte mínimo com respeito às capacidades \(u_{ij}\) contendo o menor número de arcos.

**Ford-Fulkerson:**

```
## Algoritmo de Ford-Fulkerson — Fluxo Máximo

**Entrada:** grafo G=(V,A) com capacidade u(i,j), fonte s, sorvedouro t  
**Saída:** fluxo máximo f* de s a t

---

    para cada aresta (i,j) ∈ A faça
        x(i,j) ← 0
    f* ← 0
    Construir rede residual G_R a partir de G e x

    enquanto existir caminho P de s a t em G_R faça

        // Passo 1: encontrar qualquer caminho de s a t via DFS
        P ← BuscaEmProfundidade(G_R, s, t)

        // Passo 2: calcular gargalo do caminho
        δ ← min { u_R(i,j) : (i,j) ∈ P }

        // Passo 3: enviar δ unidades pelo caminho P
        para cada aresta (i,j) ∈ P faça
            se (i,j) é aresta direta então
                x(i,j) ← x(i,j) + δ
            senão                          // aresta reversa
                x(j,i) ← x(j,i) − δ
            fim-se
        fim-para

        // Passo 4: atualizar rede residual e fluxo total
        Atualizar G_R com base no novo fluxo x
        f* ← f* + δ

**Complexidade:** O(f* · |A|)
// Depende do valor do fluxo máximo — pseudopolinomial
// Cada iteração custa O(|A|) via DFS e aumenta f* em pelo menos 1
```

**Complexidade:** O(|A| · f*)

onde f* é o valor do fluxo máximo.

O algoritmo encontra qualquer caminho aumentante de s a t na rede residual (via DFS ou BFS sem critério de escolha), envia fluxo pelo gargalo e repete. O problema é que, se as capacidades forem inteiras, cada iteração aumenta o fluxo em pelo menos 1 unidade, resultando em até f* iterações. Cada busca percorre no máximo |A| arestas.

**Exemplo ilustrativo do pior caso:**

Considere o grafo clássico com 4 nós e capacidades grandes:

```
s → A : cap = 1000
s → B : cap = 1000
A → B : cap = 1
A → t : cap = 1000
B → t : cap = 1000
```

Se Ford-Fulkerson alternar entre os caminhos s→A→B→t e s→B→A→t, ele envia apenas 1 unidade por iteração, realizando 2000 iterações para um fluxo máximo de 2000. O custo explode por causa da aresta A→B de capacidade 1 sendo usada alternadamente nas duas direções.

####B. Edmonds-Karp



**Complexidade:** O(|V| · |A|²)

É uma especialização de Ford-Fulkerson que usa BFS para sempre escolher o caminho aumentante de menor número de arestas (caminho mais curto em saltos). Isso garante que a distância de s a t na rede residual nunca diminui entre iterações, limitando o número de aumentos a O(|V| · |A|), cada um custando O(|A|) com BFS.

**Exemplo:**

No mesmo grafo acima, Edmonds-Karp encontra na 1ª iteração o caminho s→A→t (2 arestas) e envia 1000 unidades. Na 2ª, s→B→t e envia 1000. Termina em apenas 2 iterações, sem o problema de alternância do Ford-Fulkerson básico.

A complexidade O(|V|·|A|²) é independente do valor do fluxo, tornando o algoritmo fortemente polinomial.


####C. Busaker & Gowen



**Complexidade:** O(θ · (|V| log|V| + |A|))

onde θ é o valor do fluxo desejado (ou fluxo máximo).

O algoritmo encontra, a cada iteração, o caminho de menor custo na rede residual usando Dijkstra com potenciais de Johnson. Envia o máximo fluxo possível por esse caminho. No pior caso realiza θ iterações (quando o gargalo é sempre 1), cada uma custando O(|V| log|V| + |A|) com heap binário.

**Exemplo (instância da questão 7):**

Com o grafo G(V,A) , o algoritmo realizou 4 iterações:

| It. | Caminho | δ | Custo/un. |
|-----|---------|---|-----------|
| 1 | s→A→D→t | 10 | 19 |
| 2 | s→B→E→t | 8 | 19 |
| 3 | s→C→E→t | 6 | 22 |
| 4 | s→B→A→D→t | 2 | 26 |

θ = 26, custo mínimo = 526. Cada iteração escolheu o caminho de menor custo disponível, garantindo otimalidade ao final.

---

 Comparativo resumido

| Algoritmo | Complexidade | Tipo de caminho | Otimiza |
|-----------|-------------|-----------------|---------|
| Ford-Fulkerson | O(f* · \|A\|) | Qualquer (DFS) | Fluxo máximo |
| Edmonds-Karp | O(\|V\| · \|A\|²) | Menor nº de arestas (BFS) | Fluxo máximo |
| Busaker & Gowen | O(θ · (\|V\| log\|V\| + \|A\|)) | Menor custo (Dijkstra) | Fluxo máximo de custo mínimo |

Ford-Fulkerson pode ser pseudopolinomial (depende de f*); Edmonds-Karp é fortemente polinomial; Busaker & Gowen é pseudopolinomial mas garante solução ótima em custo.

## Algoritmo de Busaker & Gowen — Fluxo de Custo Mínimo

**Entrada:** grafo G=(V,A) com custo c(i,j) e capacidade u(i,j), fonte s, sorvedouro t, fluxo desejado θ  
**Saída:** fluxo x de custo mínimo com valor θ

---
    para cada aresta (i,j) ∈ A faça
        x(i,j) ← 0
    F ← 0
    Construir rede residual G_R a partir de G e x

    enquanto F < θ faça

        // Passo 1: encontrar o caminho de menor custo em G_R
        P ← CaminhoMenorCusto(G_R, s, t)   // Dijkstra com potenciais

        se não existe caminho de s a t em G_R então
            retornar "fluxo máximo atingível = F"
        fim-se

        // Passo 2: calcular gargalo do caminho
        δ ← min { u_R(i,j) : (i,j) ∈ P }
        δ ← min(δ, θ − F)

        // Passo 3: enviar δ unidades pelo caminho P
        para cada aresta (i,j) ∈ P faça
            se (i,j) é aresta direta então
                x(i,j) ← x(i,j) + δ
            senão
                x(j,i) ← x(j,i) − δ
            fim-se
        fim-para

        // Passo 4: atualizar rede residual e fluxo total
        Atualizar G_R com base no novo fluxo x
        F ← F + δ

    fim-enquanto

---

**Resultado:**

    retornar x, F, Custo(x) = Σ c(i,j)·x(i,j)

### Execução do algoritmo

O algoritmo de Busaker & Gowen envia, em cada iteração, o máximo fluxo possível pelo caminho de menor custo de s a t (usando Dijkstra com potenciais de Johnson).

**Iteração 1 — s → A → D → t**
- Custo unitário: 4 + 6 + 9 = **19**
- Gargalo (capacidade mínima do caminho): min(10, 12, 17) = **10**
- Custo adicionado: 10 × 19 = **190**

**Iteração 2 — s → B → E → t**
- Custo unitário: 8 + 3 + 8 = **19**
- Gargalo: min(20, 8, 14) = **8**
- Custo adicionado: 8 × 19 = **152**

**Iteração 3 — s → C → E → t**
- Custo unitário: 7 + 7 + 8 = **22**
